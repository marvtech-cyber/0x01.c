```python
#imports
from diffusers import DDPMScheduler
import torch
from datasets import load_dataset
import torchvision.transforms as transforms
import matplotlib.pyplot as plt # Import matplotlib.pyplot
from diffusers import UNet2DModel
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
#load dataset from huggingface datasets library
dataset = load_dataset("huggan/smithsonian_butterflies_subset", split="train")
```

```python
# Define a function to display images
def show_images(images):
  plt.figure(figsize=(8, 8))
  for i in range(images.shape[0]):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i].permute(1, 2, 0))
    plt.axis('off')
  plt.show()
```

```python
image_size = 64

#Define data augmentations
preprocess = transforms.Compose(
    [
        transforms.Resize((image_size, image_size)), # Resize
        transforms.RandomHorizontalFlip(), # Randomly flip (data augmentation)
        transforms.ToTensor(), # Convert to tensor(0, 1)
        transforms.Normalize([0.5], [0.5]), # Map to (-1, 1)

    ]
)
```

```python
#create a dataloader to load data in batches
batch_size = 32

def transform(examples):
  images = [preprocess(image.convert("RGB")) for image in examples["image"]]
  return {"images": images}

dataset.set_transform(transform)
train_dataloader = torch.utils.data.DataLoader(
    dataset, batch_size = batch_size, shuffle = True
)
```

```python
#testing if batch dataloader worked
batch = next(iter(train_dataloader))
print('Shape:', batch['images'].shape, '\nBounds:', batch['images'].min().item(), 'to', batch['images'].max().item())
show_images(batch['images'][:8]*0.5 + 0.5)
```

```python
#adding noise
scheduler = DDPMScheduler(num_train_timesteps=1000, beta_start=0.001, beta_end=0.02)
timesteps = torch.linspace(0, 999, 8).long()
x = batch['images'][:8]
noise = torch.rand_like(x)
noised_x = scheduler.add_noise(x, noise, timesteps)
show_images((noised_x*0.5 + 0.5).clip(0, 1))
```

```python
# create a UNetModel
model = UNet2DModel(
    in_channels = 3, # 3 channels for RGB images
    sample_size = 64, # Specify our input size
    block_out_channels =(64, 128, 256, 512), # N channels per layer
    down_block_types = ("DownBlock2D", "DownBlock2D", "AttnDownBlock2D", "AttnDownBlock2D"),
    up_block_types = ("AttnUpBlock2D", "AttnUpBlock2D", "UpBlock2D", "UpBlock2D"),
)
# Pass a batch of data through
with torch.no_grad():
  out = model(noised_x, timestep=timesteps).sample

out.shape
```

```python
# Training code
num_epochs = 15  # number of runs through the data that should be done

# what learning rate we should use
lr = 1e-4

# Define the device
device = "cuda" if torch.cuda.is_available() else "cpu"

model = model.to(device) # the UNET Model we made

#optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=lr)

losses = [] #array to store loss values for latter plotting

#Training the model (takes time)

for epoch in range(num_epochs):
  for step, batch in enumerate(train_dataloader):

    #Load the input images
    clean_images = batch["images"].to(device)

    # Sample noise to add to the images
    noise = torch.randn(clean_images.shape).to(clean_images.device)
#Sample a random timestep for each image
    timesteps = torch.randint(0, scheduler.num_train_timesteps, (clean_images.shape[0],), device = clean_images.device).long()

    # Add noise to clean images according to timestep
    noisy_images = scheduler.add_noise(clean_images, noise, timesteps)

    #Get the model prediction for the noise
    noise_pred = model(noisy_images, timestep = timesteps, return_dict = False)[0]

    #Compare the prediction with the actual noise
    loss = F.mse_loss(noise_pred, noise)

    # Store the loss for latter plotting
    losses.append(loss.item())

    #Update the model parameters with the optimizer based on this loss
    loss.backward(loss)
    optimizer.step()
    optimizer.zero_grad()

  #plot the loss curve
  plt.plot(losses)
  plt.show()
```
